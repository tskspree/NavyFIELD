```python
from math import comb
```

```python
def get_prob(t):
    result = 0
    for i in range(0,9):
        combination = ((-1)**i) * comb(9,i) * ((9-i)**t)
        result += combination
    prob = result/(9**t)
    return prob
```

```python
for j in range(9,50):
    estimation = '{:.2%}'.format(get_prob(j))
    words = '充值 ' + str(j*100) + ' 元能得到至少一张兑换券的概率为 ' + estimation
    print(words)
```
